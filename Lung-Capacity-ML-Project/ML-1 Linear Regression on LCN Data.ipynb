```python
lcn.Smoke = lcn.Smoke.replace({'no':0 , 'yes':1})
```

```python
#step 4 -Sampling
```

```python
from sklearn.model_selection import train_test_split
```

```python
lcn_train, lcn_test= train_test_split(lcn, test_size= 0.2)
```

```python
lcn_test.shape
```

```python
lcn_train.shape
```

```python
lcn_train_x = lcn_train.iloc[: , 1::]
lcn_test_x = lcn_test.iloc[: , 1::]
```

```python
lcn_train_y = lcn_train.iloc[: , 0]
lcn_test_y = lcn_test.iloc[: , 0]
```

```python
lcn_train_y.head()
```

```python
# Step 5 - Model Building

    # Import the neccessary function from sklearn
    # Create an object of that function
    # Run the fit function and model will be crated
    # Run the prediction funtion
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linereg = LinearRegression()
```

```python
linereg.fit(lcn_train_x , lcn_train_y)
```

```python
# we will try to evaluate the model
```

```python
# 1.Rsquare
```

```python
Rsquare = linereg.score(lcn_train_x , lcn_train_y)
Rsquare
```

```python
# 2.Adjusted Rsquare
```

```python
lcn.shape 
```

```python
 N = lcn_train_x.shape[0]
 K = lcn_train_x.shape[1]
```

```python
Adj_Rsquare = 1 - (1 - Rsquare) * (N-1) / (N-K-1)
Adj_Rsquare
```

```python
linereg.coef_
```

```python
linereg.intercept_
```

```python
pred_train = linereg.predict(lcn_train_x) 
pred_test = linereg.predict(lcn_test_x) 
```

```python
err_test = lcn_test_y - pred_test
```

```python
err_test
```

```python
# 3 MSE (mean square error)
```

```python
mse = np.mean(np.square(err_test))
mse             
```

```python
# 4 RMSE ( root mean sqaure error)
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
# 5. Mean absolute percentage error )
```

```python
mape = np.mean(np.abs(err_test * 100 / lcn_test_y))
mape
```
